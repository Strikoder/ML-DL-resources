```python
from mpl_toolkits.mplot3d import Axes3D
N_numbers=10000
N_bin=20
np.random.seed(42)
x,y = np.random.multivariate_normal(
    mean=[0.0,0.0],
    cov=[[1.0,0.4],[0.4,0.25]],
    size = N_numbers
).T
fig= plt.figure()
ax = fig.add_subplot(111, projection='3d')
hist,xedges,yedges=np.histogram2d(x,y,bins=N_bins)
plt.title("3D multi var norm distro")
plt.xlabel("x axis")
plt.ylabel("y axis")
xpos,ypos=np.meshgrid(xedges[:-1],yedges[:-1])
xpos=xpos.flatten('F')
ypos=ypos.flatten('F')
zpos=np.zeros_like(xpos)

dx=0.5*np.ones_like(zpos)
dy=dx.copy()
dz=hist.flatten()

ax.bar3d(xpos,ypos,zpos,dx,dy,dz,color='r',zsort='average')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulating some data
epochs = np.arange(1, 11)  # 10 epochs
accuracy = np.random.uniform(0.8, 0.95, size=10)
precision = np.random.uniform(0.7, 0.9, size=10)
recall = np.random.uniform(0.6, 0.85, size=10)
f1_score = np.random.uniform(0.65, 0.88, size=10)

# Plotting each metric in its subplot and then all together
fig, axs = plt.subplots(5, 1, figsize=(10, 20))

# Accuracy plot
axs[0].plot(epochs, accuracy, 'r-o', label='Accuracy')
axs[0].set_title('Accuracy over Epochs')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# Precision plot
axs[1].plot(epochs, precision, 'g--^', label='Precision')
axs[1].set_title('Precision over Epochs')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Precision')
axs[1].legend()

# Recall plot
axs[2].plot(epochs, recall, 'b-.s', label='Recall')
axs[2].set_title('Recall over Epochs')
axs[2].set_xlabel('Epoch')
axs[2].set_ylabel('Recall')
axs[2].legend()

# F1 Score plot
axs[3].plot(epochs, f1_score, 'm:x', label='F1 Score')
axs[3].set_title('F1 Score over Epochs')
axs[3].set_xlabel('Epoch')
axs[3].set_ylabel('F1 Score')
axs[3].legend()

# Combined plot
axs[4].plot(epochs, accuracy, 'r-o', label='Accuracy')
axs[4].plot(epochs, precision, 'g--^', label='Precision')
axs[4].plot(epochs, recall, 'b-.s', label='Recall')
axs[4].plot(epochs, f1_score, 'm:x', label='F1 Score')
axs[4].set_title('All Metrics over Epochs')
axs[4].set_xlabel('Epoch')
axs[4].set_ylabel('Metric Value')
axs[4].legend()

plt.tight_layout()
plt.show()





```
